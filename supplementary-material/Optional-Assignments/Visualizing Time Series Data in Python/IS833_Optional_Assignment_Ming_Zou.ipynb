```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

## Explore the dataset
The unemployment rate in the US from 2000 to 2010

```python
jobs = pd.read_csv('ch5_employment.csv')
```

```python
print(jobs.head(5))
```

```python
print(jobs.dtypes)
```

```python
# Convert datastamp column to a datetime object
jobs['datestamp'] = pd.to_datetime(jobs['datestamp'])

# Set the datestamp columns as the index
jobs = jobs.set_index('datestamp')
```

```python
# Check if there are any null value in any column
print(jobs.isnull().sum())
```

## Describe the dataset with boxplots

```python
jobs.boxplot(fontsize=12, figsize=(13, 10), vert=False)
plt.show()
```

```python
print(jobs.describe())
```

The **agriculture** industry had the highest average unemployment rate. The **construction** industry is the most variable one.

## Plot all the time series

```python
ax = jobs.plot(colormap='Spectral', fontsize=12, linewidth=0.8, figsize=(13, 10))

ax.set_xlabel('Date', fontsize=15)
ax.set_ylabel('Unemployment Rate', fontsize=15)
ax.set_title('Unemployment rate of U.S. workers by industry', fontsize=15)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

ax.axvline('2001-07-01', color='blue', linestyle='--', linewidth=0.8)
ax.axvline('2008-09-01', color='blue', linestyle='--', linewidth=0.8)

plt.show()
```

## Plot monthly and yearly trends
Display the aggregate mean values of each time series.

```python
index_month = jobs.index.month
jobs_by_month = jobs.groupby(index_month).mean()

ax = jobs_by_month.plot(fontsize=12, linewidth=1, figsize=(10, 8))

ax.set_xlabel('Month', fontsize=15)
ax.set_ylabel('Mean unemployment rate', fontsize=15)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)

plt.show()
```

```python
index_year = jobs.index.year
jobs_by_year = jobs.groupby(index_year).mean()

ax = jobs_by_year.plot(fontsize=12, linewidth=1,figsize=(10, 8))

ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Mean unemployment rate', fontsize=15)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)

plt.show()
```

## Visualize the seasonality of multiple time serie


```python
# apply times series decomposition
jobs_decomp = {}
jobs_names = jobs.columns
for ts in jobs_names:
    ts_decomposition = sm.tsa.seasonal_decompose(jobs[ts])
    jobs_decomp[ts] = ts_decomposition
```

```python
# extract the seasonal values for the decomposition of each time series
jobs_seasonal = {}
for ts in jobs_names:
    jobs_seasonal[ts] = jobs_decomp[ts].seasonal
```

```python
seasonality_df = pd.DataFrame.from_dict(jobs_seasonal)
seasonality_df.index.name = None
seasonality_df.plot(subplots=True,
                   layout=(4, 4),
                   figsize=(20, 10),
                   sharey=False,
                   fontsize=8,
                   linewidth=0.3,
                   legend=False)
plt.show()
```

## Correlations between multiple time series
Manufacturnig and Durable Goods Manufacturing have the strongest positive correlation (0.95).

Government and Construction have the strongest negative correlation (-0.71).

```python
seasonality_corr = seasonality_df.corr(method='spearman')

# the clustermap of the seasonality correlation matrix
fig = sns.clustermap(seasonality_corr, annot=True, annot_kws={"size": 10}, linewidths=.4, figsize=(20, 10))
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
plt.show()
```
